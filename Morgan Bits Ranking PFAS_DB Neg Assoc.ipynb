```python
# obtain the Morgan bit structures that are negatively correlated 
# with log CM in our PFAS database
```

```python
from rdkit import DataStructs
from rdkit import Chem
from rdkit.Chem import AllChem
import numpy as np
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
from IPython.display import SVG
```

# Bit 672

```python
molecule1 = Chem.MolFromSmiles('C1=CC(=CC=C1C2(C(C(C2(F)F)(F)F)(C3=CC=C(C=C3)Br)F)F)Br') # convert SMILES to mol format and store it
molecule1
```

```python
bi1 = {}
fp1 = AllChem.GetMorganFingerprintAsBitVect(molecule1, 2, nBits=1024, bitInfo = bi1) # use the mol file to calculate Morgan fingerprint bits
fp_arr1 = np.zeros((1,)) # initialize empty numpy array
fp_arr1 = np.zeros((1,))
DataStructs.ConvertToNumpyArray(fp1, fp_arr1) # add the morgan bits to the empty numpy array
np.nonzero(fp_arr1) # Get the bits present in the molecule
```

```python
img = Draw.DrawMorganBit(molecule1,672,bi1) # Get the structure of the morgan bit 
img 
```

```python
img.save("molecule1.png")
```

```python
molecule2 = Chem.MolFromSmiles('C12(C(C(C(C1(F)F)(F)F)(C(C(C2(F)F)(F)F)(F)F)F)(F)F)F')
molecule2
```

```python
bi2 = {}
fp2 = AllChem.GetMorganFingerprintAsBitVect(molecule2, 2, nBits=1024, bitInfo = bi2)
fp_arr2 = np.zeros((1,))
DataStructs.ConvertToNumpyArray(fp2, fp_arr2)
np.nonzero(fp_arr2)
```

```python
img = Draw.DrawMorganBit(molecule2,672,bi2)
img
```

```python
img.save("molecule2.png", size=(10000, 10000))
```
