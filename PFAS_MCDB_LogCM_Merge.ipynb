```python
# merge the MC_DB and PFAS_DB to see which PFAS are present in the MC_DB
```

```python
import pandas as pd
```

```python
df_mc = pd.read_csv('df_modred_new_chems.csv') # load in MC_DB mordred data
```

```python
df_mc.head()
```

```python
df_pfas = pd.read_csv('df_mordred_PFAS_MasterList.csv') # load in PFAS_DB mordred data
```

```python
df_pfas.head()
```

```python
# merge the PFAS_DB to the log CM columns from MC_DB using the INCHIKEY as the index 
# if log CM is not empty for a chemical then that PFAS is present in MC_DB
df_PFAS_CM = pd.merge(df_pfas, df_mc[['INCHIKEY','C:M Ratio Master', 'C:M Ratio Lipid Normalized', 'Log C:M Ratio Lipid Normalized', 
                        'C:M Ratio Non-Lipid Normalized', 'Log C:M Ratio Non-Lipid Normalized']], left_on = 'INCHIKEY_x', right_on = 'INCHIKEY', how = 'left')
df_PFAS_CM
```

```python
df_PFAS_CM.to_csv('df_modred_PFAS_MasterList_LogCM.csv')
```
