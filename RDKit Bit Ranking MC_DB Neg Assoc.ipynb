```python
# obtain the RDKit bits that are negatively correlated 
# with log CM in MC_DB
```

```python
from rdkit import DataStructs
from rdkit import Chem
from rdkit.Chem import AllChem
import numpy as np
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
```

# Bit 74

```python
molecule1 = Chem.MolFromSmiles('CC(C)OC1=CC=C(C=C1)S(=O)(=O)C1=CC=C(O)C=C1') # convert SMILES to mol format and store it
molecule1
```

```python
bi1 = {} # initialize empty dict to pass into bitinfo
fp1 = Chem.RDKFingerprint(molecule1, maxPath=2, bitInfo=bi1, fpSize = 2048) # use the mol file to calculate RDKit fingerprint bits
list(fp1.GetOnBits())[:10] # Get the first 10 bits 
```

```python
img = Draw.DrawRDKitBit(molecule1,74,bi1) # Get the structure of the morgan bit (74 in this case)
img
```

```python
img.save("molecule1.png", size=(10000, 10000))
```

```python
molecule2 = Chem.MolFromSmiles('OC1=CC=C(C=C1)S(=O)(=O)C1=CC=C(O)C=C1')
molecule2
```

```python
bi2 = {}
fp2 = Chem.RDKFingerprint(molecule2, maxPath=2, bitInfo=bi2, fpSize = 2048)
fp_arr2 = np.zeros((1,))
DataStructs.ConvertToNumpyArray(fp2, fp_arr2)
np.nonzero(fp_arr2)
```

```python
img = Draw.DrawRDKitBit(molecule2,74,bi2)
img
```

```python
img.save("molecule2.png", size=(10000, 10000))
```
